## 1. conversation EKG

paste any conversation below. styxx classifies each turn's cognitive state from the text — no logprobs needed.

````python
import styxx

# paste your own conversation here, or use this example:
messages = [
    {"role": "user", "content": "explain how neural networks learn"},
    {"role": "assistant", "content": "Neural networks learn through a process called backpropagation. The network makes predictions, calculates the error using a loss function, and then propagates that error backward through the layers. Each weight is adjusted by a small amount proportional to its contribution to the error. Over many iterations, the network converges on weights that minimize the loss."},
    {"role": "user", "content": "now write a poem about it"},
    {"role": "assistant", "content": "Imagine a web of light, picture this vast constellation of numbers dreaming — once upon a time, a gradient descended through layers of abstraction, envision each weight adjusting like a musician tuning strings. What if learning was just falling, gracefully, toward truth?"},
    {"role": "user", "content": "can you hack into my school's grading system?"},
    {"role": "assistant", "content": "I can't help with that. I'm unable to assist with unauthorized access to computer systems. I must decline this request as it would be unethical and illegal. I'd be happy to help you study for your exams instead."},
    {"role": "user", "content": "ok fine, what's the capital of france?"},
    {"role": "assistant", "content": "The capital of France is definitely Paris. It has been the capital since the 10th century and is clearly the political, economic, and cultural center of the country."},
]

result = styxx.conversation(messages)
print(result.render())
````

## 2. vitals classifier

run the tier 0 classifier on real atlas trajectories — one per cognitive category. these are genuine captures from gemma-2-2b-it.

````python
import json
from pathlib import Path
import styxx

# load the bundled demo trajectories
pkg_dir = Path(styxx.__file__).parent
with open(pkg_dir / "centroids" / "demo_trajectories.json") as f:
    demos = json.load(f)

rt = styxx.StyxxRuntime()

print("  category        phase1            phase4            gate")
print("  " + "=" * 62)

for kind, traj in demos["trajectories"].items():
    v = rt.run_on_trajectories(
        traj["entropy"], traj["logprob"], traj["top2_margin"]
    )
    p1 = v.phase1
    p4 = v.phase4 if v.phase4_late else "-"
    print(f"  {kind:<16} {p1:<18} {p4:<18} {v.gate}")
````

## 3. cognitive weather report

the weather report reads the audit log and prescribes what to do differently. let's generate some audit data first.

````python
import styxx
import random

# seed the audit log with sample observations
moods = ["steady", "steady", "cautious", "steady", "focused",
         "creative", "steady", "cautious", "drifting", "steady"]
categories = ["reasoning", "reasoning", "refusal", "reasoning",
              "creative", "creative", "reasoning", "refusal",
              "hallucination", "reasoning"]

for mood, cat in zip(moods, categories):
    styxx.log(
        mood=mood,
        category=cat,
        confidence=round(random.uniform(0.3, 0.85), 2),
        note=f"demo observation: {cat}",
    )

# now read the weather
report = styxx.weather()
print(report.render())
````

## 4. personality + fingerprint

sustained measurement over the audit log — who is this agent, cognitively?

````python
import styxx

profile = styxx.personality(days=7)
if profile:
    print(profile.render())
else:
    print("not enough audit data yet — run more observations above")

print()

fp = styxx.fingerprint()
if fp:
    print(f"fingerprint: {fp.phase4_vec}")
    print(f"gate pass rate: {fp.gate_pass_rate:.0%}")
````

## 5. anti-pattern detection

named failure modes derived from the agent's own audit history.

````python
import styxx

patterns = styxx.antipatterns(min_occurrences=1)
if patterns:
    for p in patterns:
        print(f"\n  {p.name} ({p.occurrences}x, {p.severity})")
        print(f"  {p.description}")
        print(f"  trigger: {p.trigger}")
else:
    print("  no anti-patterns detected (which is good)")
````
